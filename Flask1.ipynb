Q1. What is Flask Framework? What are the advantages of Flask Framework?
Ans: Flask is a web framework that provides libraries to build lightweight web applications in python. 
Advantages
1.It is a lightweight framework that offers hassle-free development
2.Provide flexibility to the developer to experiment with their modules or architecture
3.It is suitable for small projects
4.Offers a built-in development server and fast debugger
5.Easily scalable for the applications
6.Support for secure cookies
7.It has integrated support for unit testing

Q2. Create a simple Flask application to display ‘Hello World!!’. Attach the screenshot of the output in
Jupyter Notebook.

```python
from flask import Flask
app=Flask(__name__)
@app.route('/')
def hello():
    return 'Hello World!!'

if __name__=='__main__':
    app.run()
```

Q3. What is App routing in Flask? Why do we use app routes?
Ans: App routing is the technique used to map this specific url with the associated function intended to perform some task.It is used to access some particular pages in the web application.
For example, if we would want all users to land on google home page when user types google in browser then app routes can be used.
