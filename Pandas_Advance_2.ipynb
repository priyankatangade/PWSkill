Q2. What is the difference between the functions loc and iloc in pandas.DataFrame?
Ans: The loc function is label based data selecting method which means that we have to pass the name of the row or column which we want to select.
This method includes the last element of the range passed in it.

The iloc() function is an indexed-based selecting method which means that we have to pass an integer index in the method to select a specific row/column. 
This method does not include the last element of the range passed in it


```python
df.iloc[2]
```

```python
df.loc[(df.course_name=='Big Data')]
```

Q6. What do you understand about the windows function in pandas and list the types of windows
functions?
Ans: Window functions allow us to perform an operation with a given row’s data and data from another row that 
is a specified number of rows away — this “number of rows away value” is called the window.Window functions allow
us to perform computations among the values of a specified column.
Types of Windows Function:
    1.Rolling Window Function
    2.Weighted Window Function
    3.Expanding Window Function
    4.Exponentially Weighted Window Function

Q9. Write a Python program that reads a CSV file containing categorical data and converts a specified

name, and category order, and then display the sorted data.

```python
url = 'https://raw.githubusercontent.com/araastat/BIOF085/master/data/mtcars.csv'
file_path,column_name=input("Enter comma separated values for filepath, columnname").split(",")
category_order=input("Enter comma separated value for category").split(',')
```

```python
df=pd.read_csv(file_path)
df
```
