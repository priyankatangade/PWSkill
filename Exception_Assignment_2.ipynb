Q3. What errors are defined in the ArithmeticError class? Explain any two with an example.
Ans: FloatingPointError, OverflowError and ZeroDivisionError is defined in ArithmethicError Class.

```python
#OverflowError
i=1
try:
    f=3.0**i
    for i in range(100):
        print(i,f)
        f=f**2
except OverflowError as e:
    print('overflow after',f, e)
```

```python
#ZeroDivisionError
try:
    result=5/0
except ZeroDivisionError  as e:
    print(e)
```

Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.
Ans: The LookupError Exception is the Base class for errors raised when something can't be found.The base class for the exceptions that are raised when a key or index used on a mapping or sequence is invalid. KeyError occurs when we try to access the key which does not exist in a dictionary. IndexError occurs when we try to access the element which does not exist at that index.

```python
#indexError 
list1=[1,2,3,4]
try:
 print(list1[7])
except LookupError as e:
    print(e)
```

```python
#KeyError
movies={'name':'KNPH','actor':'Hrithik','NoOfSong':5}
movies['actor1']
```

```python
student = {
  "name": "priya",
  "course": "Data Science",
  "age": 30
}

getStudentInfo = input("What info about the student do you want? ")

try:
    print(f"The value for your request is {student[getStudentInfo]}")
except KeyError as e :
    print(f"There is no parameter with the '{getStudentInfo}' key")
```
