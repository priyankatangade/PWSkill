Q1. Explain GET and POST methods.
Ans: GET is used to request a specific data from specifed resource.The query string is sent in the URL of a GET request.
/test/demo_form.html?name1=value1&name2=value2

POST method is used to send data to a server to create/update a resource. The data sent to the post method is stored in the request body of the HTTP request.
POST /test/demo.html HTTP/1.1
Host:pwskill.com
name=value1&name2=value2

Q3. Why is redirect() used in Flask?
Ans: Flask redirect is defined as a function or utility in flask which allows developer to redirect users to a specified URL and assign a specified status code.When this function is called, a response object is returned and the redirect happens to the target location with the status code.

```python
from flask import Flask,redirect,url_for
app=Flask(__name__)

@app.route("/home")
def home():
    return redirect(url_for('LoginPage'))

app.route("/LoginPage")
def LoginPage():
    return 'you are redirected to Login Page'


if __name__=='__main__':
    app.run()
    
```
